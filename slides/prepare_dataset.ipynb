```python
import skimage
import matplotlib.pyplot as plt
```

```python
im = skimage.io.imread("https://github.com/mjirik/ZDO/raw/master/objekty/dataset/01.jpg")
plt.figure(figsize=[15,10])
plt.imshow(im)
```

```python
size = [350,350]
data = [
    [100,100],
    [250,700],
    [50,650],
    [580,600],
    [740,320],
    [200,200],
    [900,700],
    [1050,400]
    # [1100,100]
]
for i, (x, y) in enumerate(data):
    imcr = im[x:x+size[0],y:y+size[1]]
    plt.figure()
    plt.imshow(imcr)
    skimage.io.imsave(f'{i:02d}.jpg', imcr)
```

```python


import pandas as pd

df = pd.DataFrame({"filename":[f'{i:02d}.jpg' for i in range(len(data))], "contain_object":[0,0,0,1,1,0,1,1]})
df
```

```python
df.to_csv("annotation.csv", index=False)
```
